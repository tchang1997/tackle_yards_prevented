```python
import pandas as pd
import numpy as np

plays = pd.read_csv("../data/nfl-big-data-bowl-2024/plays.csv")
games = pd.read_csv("../data/nfl-big-data-bowl-2024/games.csv")
len(plays)
```

```python
plays_aug = pd.merge(plays, games, how="left", on="gameId")
plays_aug
```

```python
quarter = pd.get_dummies(plays_aug["quarter"], prefix="Q", prefix_sep="").rename(columns={"Q5": "OT"}).astype(int)

down = pd.get_dummies(plays_aug["down"], prefix="down").astype(int)

YARDAGE_BINS =  [0, 3, 7, 10, 100]
YARDAGE_CATEGORIES = ["short", "medium", "long", "very_long"]
yardage = pd.get_dummies(pd.cut(plays_aug["yardsToGo"], bins=YARDAGE_BINS, right=True, include_lowest=True, labels=YARDAGE_CATEGORIES), prefix="yardage").astype(int)

time = plays_aug["gameClock"].str.split(":", expand=True)
time_remaining = pd.Series(time.loc[:, 0].astype(int) + time.loc[:, 1].astype(int) / 60., name="time")


home_team_has_the_ball = (plays_aug["possessionTeam"] == plays_aug["homeTeamAbbr"])
possession_team_score =  plays_aug["preSnapHomeScore"].where(home_team_has_the_ball, plays_aug["preSnapVisitorScore"])
defensive_team_score =  plays_aug["preSnapVisitorScore"].where(home_team_has_the_ball, plays_aug["preSnapHomeScore"])
possession_team_deficit = possession_team_score - defensive_team_score

formation = pd.get_dummies(plays_aug["offenseFormation"], prefix="formation").astype(int)

DEFENDERS_IN_BOX_BINS = [0, 4, 5, 6, 7, 8, 12]
defenders_in_box = pd.get_dummies(pd.cut(np.nan_to_num(plays_aug["defendersInTheBox"], nan=-1), bins=DEFENDERS_IN_BOX_BINS, right=True), prefix="n_def_in_box").astype(int)

pd.concat([quarter, down, yardage, time_remaining, possession_team_score, defensive_team_score, possession_team_deficit, formation, defenders_in_box], axis=1)
```

```python
play_features = test_play['play_features']
# one-hot encode quarter (1, 2, 3, 4 ,  5  - for OT),
quarter = np.zeros(5)
quarter[play_features["quarter"] - 1] = 1.

# #one-hot encode down (1, 2, 3, 4),
down = np.zeros(4)
down[play_features["down"] - 1] = 1.

# bucket yards to go (0, 3], (3, 7], (7, 10], (10, inf],
bins = [0, 3, 7, 10, 100]
yards_to_go = np.zeros(len(bins) - 1)
yards_to_go_bin = np.digitize(play_features["yardsToGo"], bins=bins)
yards_to_go[yards_to_go_bin - 1] = 1.

# clock -> minutes, 
minutes, seconds = play_features["gameClock"].split(":")
time_remaining = int(minutes) + int(seconds) / 60.

# keep scores and yard number, add deficit,
home_team_has_the_ball = play_features["possessionTeam"] == "BUF" # play_features["homeTeam"]
possession_team_score = play_features["preSnapHomeScore"] if home_team_has_the_ball else play_features["preSnapVisitorScore"]
defensive_team_score = play_features["preSnapVisitorScore"] if home_team_has_the_ball else play_features["preSnapHomeScore"]
possession_team_deficit = possession_team_score - defensive_team_score

# onehot encode offense formation (x5), keep def in box as one hot encoded (x12)
formation = np.zeros(len(formations))
formation[FORMATIONS[play_features["offenseFormation"]]] = 1.

defenders_in_box_bins =  [0, 4, 5, 6, 7, 8, 12]
defenders_in_box = np.zeros(len(defenders_in_box_bins) - 1)
defenders_in_box_index = np.digitize(np.nan_to_num(play_features["defendersInTheBox"], nan=-1), bins=defenders_in_box_bins)
defenders_in_box[defenders_in_box_index - 1] = 1.

final_features = np.concatenate([quarter, down, yards_to_go, [time_remaining, possession_team_score, defensive_team_score, possession_team_deficit], formation, defenders_in_box])
# predict play result (yards gained/lost)
final_features.shape
```
